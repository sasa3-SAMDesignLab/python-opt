```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-3,3)
f1 = lambda x: 2*x

plt.plot(x,f1(x),color='k', linestyle='-')
plt.show()
```

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
x = y = np.linspace(-2,2)
x, y = np.meshgrid(x,y)

g1 = lambda x: 2*x[0] - 3*x[1]

ax = Axes3D(plt.figure())
ax.plot_wireframe(x,y,g1([x,y]),rstride=2, cstride=2)
plt.show()
```

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = y = np.linspace(-2,2)
x,y = np.meshgrid(x,y)

g1 = lambda x: 2*x[0] - 3*x[1]

plt.contourf(x,y,g1([x,y]), cmap=plt.cm.binary)
plt.show()
```

```python
from sympy import *
x = [Symbol('x[0]'), Symbol('x[1]')]

f1 = lambda x: (x[0]+2*x[1] - 7)**2 + (2*x[0]+x[1]-5)**2

print([diff(f1(x),x[0]), diff(f1(x),x[1])])
print([[diff(f1(x),x[0],x[0]), diff(f1(x),x[0],x[1])],
       [diff(f1(x),x[1],x[0]), diff(f1(x),x[1],x[1])]])
```

```python
import numpy as np
import scipy.linalg as linalg
a = np.random.randint(-10,10,(3,2))

A = np.dot(a.T,a)
print(A)
U = linalg.cholesky(A)
print(U)
print(np.dot(U.T,U))
```
